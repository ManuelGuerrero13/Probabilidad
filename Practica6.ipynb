```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
covid = pd.read_csv("20-06-30-COVID19MEXICO - codificado.csv")

covid
```

```python
covid.EDAD.min();
```

```python
covid.EDAD.max();
```

```python
covid.EDAD.mean();
```

```python
covid.describe();
```

```python
#me da informacion del tipo de variable para cada columna y valores nulos
covid.info();
```

```python
plt.pyplot.hist(covid.EDAD, bins = 24)
plt.pyplot.title("Histograma pruebas realizadas por covid")
plt.pyplot.xlabel("edad")
plt.pyplot.ylabel("Frecuencia")
```

```python
plt.pyplot.hist(covid.TIPO_PACIENTE, bins = 3)
plt.pyplot.title("Histograma de tipos de paciente por covid")
plt.pyplot.xlabel("tipo de paciente")
plt.pyplot.ylabel("Frecuencia")
```

```python
plt.pyplot.hist(covid.SEXO, bins = 3)
plt.pyplot.title("Histograma de pruebas covid realizadas por genero")
plt.pyplot.xlabel("genero")
plt.pyplot.ylabel("Frecuencia")
```
