```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#generar datos para la distribucion normal
media = 70 #media
desv_std = 4 #Desviacion estandar
num_datos = 500

#generamos los datos de manera aleatoria
datos = np.random.normal(media,desv_std,num_datos)
datos[0:50] #mostrar los primeros 50
```

```python
## crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std,media + 3 * desv_std,num_datos)

#calcular la PDF de la distribucion normal
pdf = 1/(desv_std * np.sqrt(2*np.pi)) * np.exp(-(x-media)**2/(2*desv_std **2))
#para no imprimir toda la lista de valores (500), desplegamos los primeros 50

print("Imprimir los primeros 50 valores de x",x[0:50]," , y los primeros 50 pdf", pdf[0:50])
```

```python
#graficar histograma
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos,bins=15,alpha=0.4,color='b') #histograma de los datos
plt.title('Distribucion normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show
```

```python
#Visualizar la PDF
plt.figure(figsize=(6,4))
plt.plot(x,pdf,label='Normal:(media:{};ds:{})'.format(np.round(media,2),np.round(desv_std,2)))
plt.title('Funcion de densidad de probabilidad(PDF) - Distribucion normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos,bins=15,alpha=0.4,density=True,color='b') #histograma de los datos
plt.plot(x,pdf,'k',linewidth=2,)

plt.title('Distribucion normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(6,4))

#Parametros de la distribucion normal
media = 70
desv_std = 4
total_estudiantes = 500

# variable para calculo de la probabilidad que un estudiante pese menos de 75 kg
peso_limite = 79

#primero necesitamos calcular la probabilidad que un estudiante pses menos
# de 75 kg utilizamos la funcion de distribucion acumulativa (CDF) de la
#distribucion normal
prob_menos_75kg = norm.cdf(peso_limite,media,desv_std)
```

```python
#calculo del numero de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75kg: ",prob_menos_75kg)
print("Numero de estudiantes que pesan menos de 75kg: ",estudiantes_menos_75kg)

#graficar la distribucion normal
x = np.linspace(media - 4 * desv_std,media + 4 * desv_std,1000)
y = norm.pdf(x,media,desv_std)

plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x <= peso_limite), color='skyblue',alpha = 0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$,$\sigma=4$,x=500 ')
plt.xlabel('peso(kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))

#Parametros de la distribucion normal
media = 70
desv_std = 4
total_estudiantes = 500

# variable para calculo de la probabilidad que un estudiante pese menos de 75 kg
peso_limite = 79

#primero necesitamos calcular la probabilidad que un estudiante pses menos
# de 62 kg utilizamos la funcion de distribucion acumulativa (CDF) de la
#distribucion normal
prob_menos_79kg = norm.cdf(peso_limite,media,desv_std)

prob_mas_79kg = 1 - prob_menos_79kg

#calculo del numero de estudiantes que pesan menos de 79 kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg


print("Probabilidad de que un estudiante pese mas de 79kg: ",prob_mas_79kg)
print("Numero de estudiantes que pesan mas de 79kg: ",estudiantes_mas_79kg)

x = np.linspace(media - 4 * desv_std,media + 4 * desv_std,1000)
y = norm.pdf(x,media,desv_std)

plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x >= peso_limite), color='skyblue',alpha = 0.5)
plt.title('Probabilidad de que un estudiante pese menos de 79 kg. \n $\mu=70$,$\sigma=4$,x=500 ')
plt.xlabel('peso(kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))

#Parametros de la distribucion normal
media = 70
desv_std = 4
total_estudiantes = 500

# variable para calculo de la probabilidad que un estudiante pese menos de 75 kg
peso_limite1 = 68
peso_limite2 = 72


prob_menos_72kg = norm.cdf(peso_limite2,media,desv_std)
prob_mas_68kg = norm.cdf(peso_limite1,media,desv_std)
prob_entre =  prob_menos_72kg - prob_mas_68kg


estudiantes_entre = total_estudiantes * prob_entre


print("Probabilidad de que un estudiante este entre 68 y 72: ",prob_entre)
print("Numero de estudiantes que estan entre 68 y 72: ",estudiantes_entre)

x = np.linspace(media - 4 * desv_std,media + 4 * desv_std,1000)
y = norm.pdf(x,media,desv_std)

plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite1) & (x <= peso_limite2), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pesan entre 68 y 72:. \n $\mu=70$,$\sigma=4$,x=500 ')
plt.xlabel('peso(kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
