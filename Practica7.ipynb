```python

plt.pyplot.hist(autos.price, bins = 50)
plt.pyplot.title("Histograma: precios de los automoviles")
plt.pyplot.xlabel("precio")
plt.pyplot.ylabel("autos")
```

```python

sns.boxplot(x = "price",y="body-style",data = autos)
plt.pyplot.title("Histograma de precios y estilo del auto")
plt.pyplot.xlabel("precio")
plt.pyplot.ylabel("estilo del auto")
```

```python
sns.boxplot(x = "price",y="drive-wheels",data = autos)
plt.pyplot.title("Histograma de precios y ruedas del auto")
plt.pyplot.xlabel("precio")
plt.pyplot.ylabel("tipo de rueda")
```

```python
sns.boxplot(x = "price",y="engine-location",data = autos)
plt.pyplot.title("Histograma de precios y ubicacion del motor del auto")
plt.pyplot.xlabel("precio")
plt.pyplot.ylabel("ubicacion del motor")
```
