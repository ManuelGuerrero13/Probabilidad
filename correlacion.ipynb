```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Datos de entrada
periodos = [2018,2019,2020,2021,2022.2023]
inversion = np.array([2,3,5,4,11,5]).reshape(-1,1)
ganancias = [20,25,34,30,40,31]

#Crear y entrenar el modelo de regresion lineal
modelo = LinearRegression()
modelo.fit(inversion,ganancias)

#Predecir la ganancia para 2024
ganancia_2024 = modelo.predict([[8]])

print("La ganancia esperada para 2024 es de: ",ganancia_2024[0])

# Coeficiente de la regresion
print("Coeficiente de la pendiente (m): ",modelo.coef_[0])
print("Intercepcion de la recta (b): ", modelo.intercept_)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#Datos de entrada
publicidad = np.array([2,3,5,4,11,5]).reshape(-1,1)
ganancias = np.array([20,25,34,30,40,31])

#Crear y entrenar el modelo de regresion lineal
modelo = LinearRegression()
modelo.fit(publicidad,ganancias)

#Predecir la ganancia para una inversion de 8 en publicidad
inversion_prediccion = np.array([[8]])
ganancia_predicha = modelo.predict(inversion_prediccion)

print("La ganancia esperada para una inversion de 8 en publicidad es de: ", ganancia_predicha[0])
print("valor de la intercepcion", modelo.intercept_)
print("valor de la pendiente ",modelo.coef_)

#Crear el diagrama de dispersion de la linea de regresion
plt.figure(figsize=(6,5))
plt.scatter(publicidad,ganancias,color='blue',label="Datos")
plt.plot(publicidad,modelo.predict(publicidad),color='red')
plt.scatter(inversion_prediccion,ganancia_predicha,color='green',marker='x',s=100,
            label='prediccion (8,{0:2f})'.format(ganancia_predicha[0]))
plt.title('Pronostico de ganancias a partir de la inversion en publicidad')
plt.xlabel('Inversion en publicidad (Millones)')
plt.ylabel('ganancias anuales(Millones)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

#Datos de entrada
publicidad = [2,3,5,4,11,5]
ganancias = [20,25,34,30,40,31]

#Calcular el coeficiente de correlacion de pearson
coeficiente_correlacion,p_valor = pearsonr(publicidad,ganancias)

print("coeficiente de correlacion de pearson: ",coeficiente_correlacion)
print("valor p: ",p_valor)
```
